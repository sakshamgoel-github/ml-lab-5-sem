```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv("Pokemon.csv")
```

```python
df.head()
```

```python
df["Strength"].unique()
```

```python
strength = ['Strong','Medium','Low']
```

```python
enc = OrdinalEncoder(categories=[strength])
```

```python
enc.fit_transform(df[["Strength"]])
```

```python
df[["Strength"]]=enc.fit_transform(df[["Strength"]])
```
