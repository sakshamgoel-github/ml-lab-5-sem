```python
import pandas as pd
import numpy as numpy
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv("Pokemon.csv")
```

```python
df.isnull().sum().sum()
```

```python
df.head()
```

```python
df["Type"].unique()
```

```python
enc = OneHotEncoder()
```

```python
enc.fit_transform(df[["Type"]]).toarray()
```

```python
feature_array = enc.fit_transform(df[["Type"]]).toarray()
```

```python
enc.categories_
```

```python
feature_labels = enc.categories_
```

```python
pd.DataFrame(feature_array,columns=feature_labels)
```

```python
final_features = pd.DataFrame(feature_array,columns=feature_labels)
```

```python
final_features.head()
```

```python
final_df=pd.concat([df,final_features],axis=1)
```
